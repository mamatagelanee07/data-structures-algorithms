## Problem Statement
Write a function that determines if two strings are anagrams of each other. Two strings are anagrams if they contain the same characters with the same frequencies, just rearranged.

### Inputs
- s: the first string
- t: the second string
### Output
true if the strings are anagrams, false otherwise

### Examples
- "listen" and "silent" are anagrams
- "hello" and "world" are not anagrams
- "rat" and "car" are not anagrams

### Constraints
- `0 <= s.length, t.length <= 10^5`
- `s` and `t` consist of printable ASCII characters (`' '` to `'~'`).
- Comparison is **case-insensitive** (`"Listen"` and `"silent"` are anagrams).
- Characters are counted exactly as-is after case normalization (spaces and punctuation are included).
- Let `n = s.length` and `m = t.length`; target complexity is `O(n + m)` time and `O(k)` extra space, where `k` is the number of distinct normalized characters.


## Approach #1 : Two Hash Map (Frequency Count)

```kotlin
fun isAnagramTwoMap(s : String, t: String) : Boolean {
    if (s.length != t.length) return false
    val sMap = mutableMapOf<Char, Int>()
    val tMap = mutableMapOf<Char, Int>()
    for (char in s){
        sMap[char.lowercaseChar()] = sMap.getOrDefault(char.lowercaseChar(), 0) + 1
    }
    for(char in t){
        tMap[char.lowercaseChar()] = tMap.getOrDefault(char.lowercaseChar(), 0) +1
    }
    return sMap.equals(tMap)
}
println(isAnagramTwoMap("listen", "silent"))
println(isAnagramTwoMap("Listen", "silent"))
println(isAnagramTwoMap("rat", "car"))
println(isAnagramTwoMap("", "car"))
```

### Developer Notes
- To determine if two strings are anagram, it's basically look at the frequency of each character in both strings and compare them. If they match, then the strings are anagrams.
- We can use a hash map (or dictionary) to count the frequency of each character in both strings. We can iterate through each character in the first string, convert it to lowercase, and update the count in the hash map. We can do the same for the second string. Finally, we can compare the two hash maps. If they are equal, then the strings are anagrams; otherwise, they are not.

### Time and Space Complexity
#### Time Complexity : O(n + m)
- The time complexity of this approach is O(n + m), where n and m are the lengths of the two strings, because we need to iterate through both strings once. The space complexity is O(k), where k is the number of distinct normalized characters, because we are storing the frequency of each character in a hash map.
#### Space Complexity : O(k)
- The space complexity is O(k), where k is the number of distinct normalized characters, because we are storing the frequency of each character in a hash map. In the worst case, if all characters are distinct, the space complexity would be O(n + m) where n and m are the lengths of the two strings. However, since we are only counting characters, the space complexity is generally considered O(k) where k is the number of distinct characters, which is typically much smaller than n and m.

## Approach #2 : One Hash Map (Frequency Count)

```kotlin
fun isAnagramOneMap(s : String, t: String) : Boolean {
    if (s.length != t.length) return false
    val map = mutableMapOf<Char, Int>()
    for (index in 0..s.length-1){
        val charS = s[index].lowercaseChar()
        val charT = t[index].lowercaseChar()
        map[charS] = map.getOrDefault(charS, 0) + 1
        map[charT] = map.getOrDefault(charT, 0) - 1
    }
    return map.values.all { it == 0 }
}
println(isAnagramOneMap("listen", "silent"))
println(isAnagramOneMap("Listen", "silent"))
println(isAnagramOneMap("rat", "car"))
println(isAnagramOneMap("", "car"))
```

### Developer Notes
- This approach uses a **single hash map** to track frequency balance between `s` and `t`.
- After confirming `s.length == t.length`, we iterate once through both strings at the same index:
  - increment count for `s[i].lowercaseChar()`
  - decrement count for `t[i].lowercaseChar()`
- Because comparison is case-insensitive, both characters are normalized with `lowercaseChar()` before counting.
- If the strings are anagrams, every character’s net count ends at `0`; otherwise, at least one count is non-zero.
- This method is more space-efficient in practice than building two separate maps, while keeping the same asymptotic complexity.

### Time and Space Complexity
#### Time Complexity: O(n + m)
- Length check is O(1), and the main loop touches each character once.
- With the early length guard, effective work is linear in input size.

#### Space Complexity: O(k)
- `k` is the number of distinct normalized characters stored in the map.
- Under printable ASCII constraints, `k` is bounded (at most 95), so auxiliary space is effectively constant in practice.

## Gotchas (Flash Cards)

### 1) Length mismatch
- **Gotcha:** If `s.length != t.length`, they cannot be anagrams.
- **Quick check:** Return `false` immediately before any counting.
- **Flash test:** `("", "car") -> false`

### 2) Case-insensitive requirement
- **Gotcha:** `"Listen"` vs `"silent"` should be true only if you normalize case.
- **Quick check:** Convert each character with `lowercaseChar()` while counting.
- **Flash test:** `("Listen", "silent") -> true`

### 3) Spaces and punctuation are valid characters
- **Gotcha:** Do not remove spaces/symbols unless problem explicitly says to ignore them.
- **Quick check:** Count `' '`, `'!'`, etc. like normal chars.
- **Flash test:** `("a b!", "!ba ") -> true`

### 4) Empty strings
- **Gotcha:** Two empty strings are anagrams.
- **Quick check:** With length check + counting logic, this should naturally return `true`.
- **Flash test:** `("", "") -> true`

### 5) Frequency, not just presence
- **Gotcha:** Same letters are not enough; counts must match exactly.
- **Quick check:** Verify per-character final count is zero (not just total sum).
- **Flash test:** `("aab", "abb") -> false`

### 6) One-map validation bug
- **Gotcha:** `map.values.sum() == 0` can be wrong.
- **Quick check:** Use `map.values.all { it == 0 }`.
- **Flash test:** `("ab", "cc") -> false` (sum is 0 but not anagrams)

### 7) Loop bounds safety
- **Gotcha:** `0..s.length-1` is harder to read and error-prone.
- **Quick check:** Prefer `for (i in s.indices)` after length check.
- **Flash test:** `("a", "a") -> true`

### 8) Complexity wording
- **Gotcha:** You loop once, but process both strings per iteration.
- **Quick check:** Say `O(n)` when `n == m`, or `O(n + m)` in general two-input notation.
- **Flash line:** "After length check, it is linear."
