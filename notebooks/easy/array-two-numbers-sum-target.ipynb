## Problem Statement
You are given an array of integers nums and an integer value target. Your task is to find two numbers in the array that add up to the target value and return their indices.

### Inputs
- nums: unsorted array of integers
- target: integer
### Output
- array : indices of two numbers in array adding up to target value

### Examples
- nums = [2, 7, 11, 15], target = 9 → [0,1]
- nums = [4,1,2,1,3], target = 7 → [0,4]

### Constraints
- 1 <= nums.length <= 3 * 10^4
- there's exactly one pair of numbers that sum to the target


## Approach #1 : Two Loops

```kotlin
fun twoSumBruteForce(nums: IntArray, target: Int): IntArray {
    println("Input: ${nums.contentToString()}, $target")
    val list = mutableListOf<Int>()
    for (i in nums.indices) {
        for (j in i+1..nums.lastIndex) {
            if(nums[j] == target - nums[i]) {
                list.add(i)
                list.add(j)
                return list.toIntArray()
            }
        }
    }
    return list.toIntArray()
}
println("${twoSumBruteForce(intArrayOf(2, 7, 11, 15), 9).contentToString()}")
println("${twoSumBruteForce(intArrayOf(4, 1, 2, 1, 3), 7).contentToString()}")
```

## Approach#2 : Hash Table

```kotlin
fun twoSumHashMap(nums: IntArray, target: Int): IntArray {
    println("Input: ${nums.contentToString()}, $target")
    val seen = HashMap<Int, Int>() // value -> index

    for (i in nums.indices) {
        val need = target - nums[i]
        val j = seen[need]
        if (j != null) return intArrayOf(j, i) // previous index first
        seen[nums[i]] = i
    }

    // Constraints say one solution exists; keep fallback for safety.
    return intArrayOf()
}
println("${twoSumHashMap(intArrayOf(2, 7, 11, 15), 9).contentToString()}")
println("${twoSumHashMap(intArrayOf(4, 1, 2, 1, 3), 7).contentToString()}")
```

## Approach#3 : Sorted + Two Pointers

```kotlin
fun twoSumSortedTwoPointers(nums: IntArray, target: Int): IntArray {
    println("Input: ${nums.contentToString()}, $target")
    val map = nums.mapIndexed { index, i -> index to i }.sortedBy { it.second }
    var left = 0
    var right = map.lastIndex
    while (left < right) {
        val leftV = map[left]
        val rightV = map[right]
        val sum = leftV.second + rightV.second
        when{
            target == sum -> return intArrayOf(leftV.first, rightV.first)
            sum < target -> left++
            else -> right--
        }
    }

    // Constraints say one solution exists; keep fallback for safety.
    return intArrayOf()
}
println("${twoSumSortedTwoPointers(intArrayOf(2, 7, 11, 15), 9).contentToString()}")
println("${twoSumSortedTwoPointers(intArrayOf(4, 1, 2, 1, 3), 7).contentToString()}")
```

### Developer Notes

#### Approach #1: Two Loops (Brute Force)
- Iterate `i` from left to right.
- For each `i`, scan `j` from `i + 1` to end.
- If `nums[i] + nums[j] == target`, return `[i, j]`.

#### Why this works
- It checks every unique pair exactly once.
- Since prompt guarantees exactly one valid pair, first match is sufficient.

#### Notes on implementation
- Simple and reliable baseline.
- No extra data structure needed.
- Good for understanding before optimizing.

---

#### Approach #2: Hash Table (One Pass)
- Maintain a map: `value -> index` for numbers seen so far.
- For current number `x`, compute `need = target - x`.
- If `need` already exists in map, return `[map[need], currentIndex]`.
- Otherwise store current value/index and continue.

#### Why this works
- At index `i`, map contains only prior elements (`0..i-1`).
- So when complement is found, indices are guaranteed distinct and valid.

#### Notes on implementation
- This is the optimal interview solution for standard Two Sum constraints.
- Requires extra memory for the map.
- Prefer checking map **before** inserting current value.

---

#### Approach #3: Sorted + Two Pointers
- Build pairs as `(originalIndex, value)`.
- Sort pairs by `value`.
- Use two pointers (`left`, `right`) to scan for sum:
  - if sum == target -> return original indices from the pairs
  - if sum < target -> move `left++`
  - else -> move `right--`

#### Why this works
- Sorting enables monotonic pointer movement.
- Two pointers examine candidate sums efficiently in one pass after sorting.

#### Notes on implementation
- Must preserve original indices before sorting.
- Good pattern practice, but not asymptotically faster than hash map for this problem.

## Time and Space Complexity

### Approach #1: Two Loops
- **Time:** `O(n^2)`
  - Nested loop over pairs.
- **Space:** `O(1)` extra
  - No auxiliary structure proportional to input size.

### Approach #2: Hash Table (One Pass)
- **Time:** `O(n)` average
  - Each lookup/insert in hash map is average `O(1)`.
- **Space:** `O(n)`
  - Map may store up to `n` elements.

### Approach #3: Sorted + Two Pointers
- **Time:** `O(n log n)`
  - Sorting dominates; pointer scan is `O(n)`.
- **Space:** `O(n)`
  - Stores `(index, value)` pairs to preserve original indices.

## Gotchas (Flash Cards)

### 1) Hash map insertion order matters
- **Gotcha:** If you insert current element before checking complement, you can accidentally reuse same index in some variants.
- **Quick check:** `check complement first`, then `put current`.

### 2) Return index order consistency
- **Gotcha:** Mixed order (`[current, previous]` vs `[previous, current]`) can fail strict tests.
- **Quick check:** Return `[map[need], i]` consistently.

### 3) Duplicate values
- **Gotcha:** Arrays like `[3,3]`, target `6` must return two different indices.
- **Quick check:** One-pass map handles this correctly when check-before-put is used.

### 4) Range boundaries in brute force
- **Gotcha:** `for (j in i+1..nums.lastIndex)` is fine only when `i < lastIndex`.
- **Quick check:** Works with `i` over `nums.indices`, but keep an eye on readability; `until` can be clearer.

### 5) Preserve original indices in sorted approach
- **Gotcha:** Sorting raw values loses original index positions.
- **Quick check:** Sort `(index, value)` pairs, not values alone.

### 6) Returning pointer positions by mistake
- **Gotcha:** In sorted two-pointer, returning `[left, right]` is wrong.
- **Quick check:** Return `[leftV.first, rightV.first]`.

### 7) Complexity misconception for sorted approach
- **Gotcha:** Two-pointer scan is linear, but total is not `O(n)` because sorting is required.
- **Quick check:** Final complexity is `O(n log n)`.

### 8) Fallback return behavior
- **Gotcha:** Prompt says exactly one solution, but generic code should still have safe fallback.
- **Quick check:** Return `intArrayOf()` (or throw) if no pair is found, and document choice.
