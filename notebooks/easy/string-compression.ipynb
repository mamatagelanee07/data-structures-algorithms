## Problem Statement
Write a function that compresses a string by replacing consecutive identical characters with the character followed by the count of repetitions.

### Inputs
- s: string to compress
### Output
- s: compressed string

### Examples
- s = "aaabbc" -> "a3b2c1"
- s = "aabbccdd" -> "a2b2c2d2"
- s = "abc" -> "a1b1c1"

### Constraints
- 1 < s.length <= 10^5


## Approach #1 : Two Pointers

```kotlin
fun compressString(s: String): String {
    println("input : $s")
    val final = StringBuilder()
    var current = 0
    var counter = 1
    while (counter < s.length){
        if (s[current] != s[counter]){
            final.append("${s[current]}${counter-current}")
            current = counter
        }
        if (counter == s.length - 1){
            final.append(s[counter])
            final.append("${counter-current+1}")
        }
        counter++
    }
    return final.toString()
}
println("Output : ${compressString("aaabbc")}")
println("Output : ${compressString("aabbccdd")}")
println("Output : ${compressString("abc")}")
println("Output : ${compressString("ab")}")
```

## Approach #2 : Single Pass + Last Group Flush

```kotlin
fun compressStringV2(s: String): String {
    if (s.isEmpty()) return ""

    val result = StringBuilder()
    var count = 1

    for (i in 1 until s.length) {
        if (s[i] == s[i - 1]) {
            count++
        } else {
            result.append(s[i - 1]).append(count)
            count = 1
        }
    }

    // Flush final character group
    result.append(s[s.length - 1]).append(count)
    return result.toString()
}

println("V2 Output : ${compressStringV2("aaabbc")}")    // a3b2c1
println("V2 Output : ${compressStringV2("aabbccdd")}")  // a2b2c2d2
println("V2 Output : ${compressStringV2("abc")}")       // a1b1c1
println("V2 Output : ${compressStringV2("ab")}")        // a1b1
println("V2 Output : ${compressStringV2("aaaaa")}")     // a5
```

### Developer Notes

#### Approach #1: Two Pointers (Run-Length Encoding)
- We keep two indices:
  - `current`: start index of the current character run
  - `counter`: scans forward through the string
- When `s[current] != s[counter]`, the current run ends:
  - append `s[current]`
  - append run length: `counter - current`
  - move `current = counter`
- When `counter` reaches the final index, append the last run as well.

#### Why this works
- Every run of consecutive identical characters is emitted exactly once.
- Run length is calculated from index distance, so counts are accurate.
- The scan moves left to right without reprocessing characters.

#### Time and Space Complexity
- **Time Complexity:** `O(n)`
  - `counter` moves from left to right once.
- **Space Complexity:** `O(n)`
  - output uses `StringBuilder` and can grow proportionally to input size.

#### Edge Cases to Validate
- Two chars same: `"aa" -> "a2"`
- Two chars different: `"ab" -> "a1b1"`
- All unique: `"abc" -> "a1b1c1"`
- Single run: `"aaaaa" -> "a5"`
- Mixed runs: `"aaabbc" -> "a3b2c1"`
