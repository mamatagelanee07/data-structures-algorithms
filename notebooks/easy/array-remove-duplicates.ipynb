## Problem Statement
Given an integer array nums sorted in non-decreasing order, remove the duplicates in-place such that each unique element appears only once. The relative order of the elements should be kept the same.
Consider the number of unique elements in nums to be k. After removing duplicates, return the number of unique elements k.
The first k elements of nums should contain the unique numbers in sorted order. The remaining elements beyond index k - 1 can be ignored.

### Inputs
- nums: array of integers sorted in non-decreasing order
### Output
- k representing number of unique elements in the modified array

### Examples
- nums = [1,1,2] -> k = 2, nums = [1,2,_]
- nums = [0,0,1,1,1,2,2,3,3,4] -> k = 5, nums = [0,1,2,3,4,_,_,_,_,_]

### Constraints
- 1 <= nums.length <= 3 * 10^4
- -100 <= nums[i] <= 100
- nums is sorted in non-decreasing order.


## Approach #1 : Two Pointers(Read & Write)

```kotlin
fun removeDuplicates(nums : IntArray) : Pair<Int, IntArray> {
    if (nums.size < 2) return Pair(nums.size, nums)
    var write = 0
    for(read in 1 until nums.size) {
        if (nums[read] != nums[write]){
            write++
            nums[write] = nums[read]
        }
    }
    return write+1 to nums
}
val output1 = removeDuplicates(intArrayOf(1,1,2))
val output2 = removeDuplicates(intArrayOf(0,0,1,1,1,2,2,3,3,4))
println("${output1.first} & [${output1.second.joinToString(",")}]")
println("${output2.first} & [${output2.second.joinToString(",")}]")
```

### Developer Notes
- Two pointers, read & write, read will go through each element in array/list & compare it with element at write index, if both are unique, then move write to next & write down element at that index. If not same, move on to reading next element. So basically it's rebuilding list instead of removing similar elements.

### Time and Space Complexity
#### Time Complexity : O(n)
- The time complexity of this approach is O(n), given we will have to traverse at least once for all elements.
#### Space Complexity : O(1)
- The space complexity is O(1) given we are not adding any extra memory & reusing the same array

## Gotchas (Flash Cards)

### 1) The array is already sorted
- **Gotcha:** This solution works because duplicates are adjacent.
- **Quick check:** If the array were unsorted, this exact two-pointer approach would not be enough.
- **Flash test:** `[2,1,2]` is not a valid input for this problem.

### 2) Return `k`, not a new array
- **Gotcha:** The problem asks you to modify `nums` in-place and return the number of unique elements.
- **Quick check:** Your function should return `Int`, not `Pair`, `List`, or a new array.
- **Flash test:** `[1,1,2] -> return 2`

### 3) Only the first `k` elements matter
- **Gotcha:** Values after index `k - 1` do not matter.
- **Quick check:** Do not waste time trying to "clean up" the rest of the array.
- **Flash test:** `[1,2,2]` with `k = 2` is acceptable because only first 2 positions matter.

### 4) First element is always unique
- **Gotcha:** For a non-empty sorted array, the first element should always stay.
- **Quick check:** Initialize `write = 0`, then start reading from index `1`.
- **Flash test:** `[5] -> k = 1`

### 5) Compare with the last unique element
- **Gotcha:** Compare `nums[read]` with `nums[write]`, not just `nums[read - 1]`.
- **Quick check:** `write` marks the last unique value in the compacted prefix.
- **Flash test:** In `[1,1,2,2,3]`, compare against the current unique tail.

### 6) Do not shift elements one-by-one on every duplicate
- **Gotcha:** Repeated shifting makes the solution unnecessarily complex and can lead to O(n²).
- **Quick check:** Just overwrite the next unique position using `write`.
- **Flash line:** "Compact the array; don't delete from it."

### 7) Watch for off-by-one errors
- **Gotcha:** The number of unique elements is `write + 1`, not just `write`.
- **Quick check:** If `write` stores the last unique index, count is index + 1.
- **Flash test:** `[1,2] -> write ends at 1, so k = 2`

### 8) Handle small input sizes correctly
- **Gotcha:** Arrays of length `1` should return `1`.
- **Quick check:** If `nums.size < 2`, return `nums.size`.
- **Flash test:** `[7] -> 1`

### 9) All duplicates case
- **Gotcha:** If all numbers are the same, result should still keep one copy.
- **Quick check:** `write` never moves, but answer is still `1`.
- **Flash test:** `[2,2,2,2] -> k = 1`

### 10) Already unique case
- **Gotcha:** If there are no duplicates, array should remain unchanged.
- **Quick check:** `write` keeps moving with `read`.
- **Flash test:** `[1,2,3,4] -> k = 4`

### 11) Preserve relative order
- **Gotcha:** You cannot reorder elements.
- **Quick check:** Unique values must stay in sorted order as they originally appear.
- **Flash test:** `[0,0,1,1,2] -> [0,1,2,...]`

### 12) In-place does not mean using extra storage
- **Gotcha:** Avoid using a `Set`, extra list, or another array unless interviewer explicitly allows it.
- **Quick check:** Only use a few variables like `read` and `write`.
- **Flash line:** "O(1) extra space."
