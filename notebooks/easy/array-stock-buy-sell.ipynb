## Problem Statement
You are given an integer array prices where prices[i] is the price of a given stock on the ith day.
On each day, you may decide to buy and/or sell the stock. You can only hold at most one share of the stock at any time. However, you can sell and buy the stock multiple times on the same day, ensuring you never hold more than one share of the stock.

Find and return the maximum profit you can achieve with prices.

### Inputs
- nums: array of integers sorted in non-decreasing order
### Output
- k representing number of unique elements in the modified array
- nums: array of integers sorted in non-decreasing order

### Examples
- prices = [7,1,5,3,6,4] -> 7
- prices = [1,2,3,4,5] -> 4
- prices = [7,6,4,3,1] -> 0

### Constraints
- 1 <= prices.length <= 3 * 10^4
- 0 <= prices[i] <= 10^4

## Approach #1 : Two Pointers(Today, Yesterday)

```kotlin
fun makeProfitBackwards(prices : IntArray) : Pair<Int, IntArray> {
    var profit = 0
    for(today in 1 until prices.size) {
        if (prices[today] > prices[today-1]) {
            profit += prices[today] - prices[today - 1]
        }
    }
    return profit to prices
}
val output1 = makeProfitBackwards(intArrayOf(7,1,5,3,6,4))
val output2 = makeProfitBackwards(intArrayOf(1,2,3,4,5))
val output3 = makeProfitBackwards(intArrayOf(7,6,4,3,1))
println("Profit is ${output1.first} for [${output1.second.joinToString(",")}]")
println("Profit is ${output2.first} for [${output2.second.joinToString(",")}]")
println("Profit is ${output3.first} for [${output3.second.joinToString(",")}]")
```

## Approach #2 : Two Pointers (Today, Tomorrow)

```kotlin
fun makeProfitForward(prices: IntArray): Pair<Int, IntArray> {
    var profit = 0

    for (today in 0 until prices.size - 1) {
        val tomorrow = today + 1
        if (prices[tomorrow] > prices[today]) {
            profit += prices[tomorrow] - prices[today]
        }
    }

    return profit to prices
}

val output4 = makeProfitForward(intArrayOf(7,1,5,3,6,4))
val output5 = makeProfitForward(intArrayOf(1,2,3,4,5))
val output6 = makeProfitForward(intArrayOf(7,6,4,3,1))

println("Profit is ${output4.first} for [${output4.second.joinToString(",")}]")
println("Profit is ${output5.first} for [${output5.second.joinToString(",")}]")
println("Profit is ${output6.first} for [${output6.second.joinToString(",")}]")
```

### Developer Notes
- This approach compares each day's price with the next day's price.
- If tomorrow's price is higher than today's, we add that increase to the total profit.
- This works because the problem allows multiple transactions, as long as we hold at most one stock at a time.
- Summing all positive adjacent differences gives the same maximum profit as buying at each valley and selling at each peak.
- This is equivalent to the `today & yesterday` approach; it just looks forward instead of backward.

### Time and Space Complexity
#### Time Complexity: O(n)
- We iterate through the prices array once.

#### Space Complexity: O(1)
- We use only a few extra variables.

## Gotchas (Flash Cards)

### 1) This problem allows multiple transactions
- **Gotcha:** This is **not** the “buy once, sell once” version.
- **Quick check:** You can take profit from multiple upward moves.
- **Flash test:** `[7,1,5,3,6,4] -> 7`, not `5`

### 2) You can only hold one stock at a time
- **Gotcha:** You cannot stack multiple buys before selling.
- **Quick check:** But you *can* sell and buy again on later days.
- **Flash line:** "One stock at a time, many transactions allowed."

### 3) Add only positive differences
- **Gotcha:** Only profit from rises; ignore drops.
- **Quick check:** `if (prices[i] > prices[i - 1]) profit += prices[i] - prices[i - 1]`
- **Flash test:** `[7,6,4,3,1] -> 0`

### 4) Continuous rise can be split into daily gains
- **Gotcha:** `1 -> 2 -> 3 -> 4 -> 5` is same as buying at `1` and selling at `5`
- **Quick check:** `(2-1) + (3-2) + (4-3) + (5-4) = 4`
- **Flash test:** `[1,2,3,4,5] -> 4`

### 5) No need to track explicit buy/sell state
- **Gotcha:** Using `stocks`, `buyPrice`, or peak/valley tracking can overcomplicate this version.
- **Quick check:** Summing positive adjacent differences is enough.
- **Flash line:** "Every climb counts."

### 6) Today/Yesterday and Today/Tomorrow are equivalent
- **Gotcha:** Both approaches check adjacent pairs; they just index differently.
- **Quick check:** `prices[i] > prices[i-1]` and `prices[i+1] > prices[i]` cover the same rises.
- **Flash line:** "Backward or forward, same daily gaps."

### 7) Watch the loop bounds in today/tomorrow
- **Gotcha:** If you use `today + 1`, don’t run to the last index.
- **Quick check:** Use `for (today in 0 until prices.size - 1)`
- **Flash test:** Avoid out-of-bounds on last day.

### 8) Single day means no profit
- **Gotcha:** You need at least 2 days to make a transaction.
- **Quick check:** Arrays of length `1` should return `0`
- **Flash test:** `[5] -> 0`

### 9) Flat prices give no profit
- **Gotcha:** Equal prices do not add anything.
- **Quick check:** Use `>` not `>=`
- **Flash test:** `[2,2,2,2] -> 0`

### 10) Return just the profit
- **Gotcha:** This problem does not require modifying the array.
- **Quick check:** Clean signature is `fun maxProfit(prices: IntArray): Int`
- **Flash line:** "Return total profit, not a pair."

### 11) Greedy is optimal here
- **Gotcha:** You do not need DP for this version.
- **Quick check:** Since every rise can be captured independently, greedy works.
- **Flash line:** "Take every positive step."

### 12) Don’t confuse this with Stock I
- **Gotcha:** In Stock I, only one transaction is allowed; here multiple are allowed.
- **Quick check:** Stock I for `[7,1,5,3,6,4]` is `5`, but this problem’s answer is `7`
- **Flash test:** Same input, different problem rules
