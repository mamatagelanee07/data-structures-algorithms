## Problem Statement
You are given the head of a singly linked list. Your task is to reverse the entire linked list and return the head of the reversed list.

### Inputs
- head: node representing head of the linked list
### Output
- head : head of reversed linked list

### Examples
- head = [1,2,3,4,5] → [5,4,3,2,1]
- head = [1,2] → [2,1]
- head = [] → []

### Constraints
- The number of nodes in the list is the range [0, 5000].
- -5000 <= Node.val <= 5000


```kotlin
data class ListNode(var `val`: Int, var next: ListNode? = null)

fun buildLinkedList(values: List<Int>): ListNode? {
    if (values.isEmpty()) return null
    val head = ListNode(values[0])
    var current: ListNode = head
    for (i in 1 until values.size) {
        val node = ListNode(values[i])
        current.next = node
        current = node
    }
    return head
}

fun toList(head: ListNode?): List<Int> {
    val out = mutableListOf<Int>()
    var current : ListNode? = head
    while (current != null) {
        val node = current
        node?.let {
            out.add(node.`val`)
            current = node.next
        }

    }
    return out
}

val head1 = buildLinkedList(listOf(1,2,3,4,5))
val head2 = buildLinkedList(listOf(1,2))
val head3 = buildLinkedList(listOf(1))
```

## Approach #1 : Arranging a paper stack

```kotlin
fun reverseList(head : ListNode?) : ListNode? {
    println("Input: ${toList(head)}")
    if (head == null) return null
    if(head.next == null) return head
    var dummy = ListNode(head.`val`)
    var current = head
    while(current != null){
        val next = current.next
        current.next = dummy.next
        dummy.next = current
        current = next
    }
    return dummy.next
}
println("${toList(reverseList(head1))}")
println("${toList(reverseList(head2))}")
println("${toList(reverseList(head3))}")
```

### Developer Notes

#### Approach #1: Dummy Node (Paper Stack)
- Create a dummy node to act as a "working head".
- Traverse the original list from left to right.
- For each node, insert it at the front of the dummy's chain:
  - Save next node: `val next = current.next`
  - Link current to dummy's next: `current.next = dummy.next`
  - Update dummy to point to current: `dummy.next = current`
  - Move forward: `current = next`
- Return `dummy.next` (the new reversed head).

#### Why this works
- The dummy node acts like a "stack" where each node is pushed to the front.
- By always inserting at the front, we reverse the order naturally.
- No pointer swapping complexity; just link rewiring.

#### Notes on implementation
- Dummy node starts empty (`next = null`).
- Original head becomes the tail of the reversed list.
- Works for all edge cases (empty, single node, multiple nodes).

## Time and Space Complexity

### Approach #1: Dummy Node
- **Time:** `O(n)`
  - Single pass through all nodes.
- **Space:** `O(1)` extra
  - Only one dummy node; no array/map/recursion stack.

## Gotchas (Flash Cards)

### 1) Dummy node initialization
- **Gotcha:** Dummy should start with `next = null`.
- **Quick check:** `val dummy = ListNode(0)` with implicit `next = null`.
- **Flash test:** Empty list reversal should return `null`.

### 2) Return `dummy.next`, not `dummy`
- **Gotcha:** Returning `dummy` itself includes the placeholder value.
- **Quick check:** Always return `dummy.next`.
- **Flash line:** "Dummy is invisible; return its next."

### 3) Order of pointer updates matters
- **Gotcha:** Must save `current.next` before changing `current.next`.
- **Quick check:** First line in loop: `val next = current.next`.
- **Flash test:** Forgetting this causes infinite loops or lost nodes.

### 4) Original head becomes the tail
- **Gotcha:** After reversal, the original head node now points to `null`.
- **Quick check:** This is correct behavior; the list is reversed.
- **Flash line:** "First node ends as last node."

### 5) Works on already-reversed lists
- **Gotcha:** If you call `reverseList` twice, original order is restored.
- **Quick check:** This is mathematically correct; double reverse = identity.
- **Flash test:** `reverseList(reverseList(head)) == head` structurally.

### 6) Null safety in Kotlin
- **Gotcha:** `head?.let { ... }` syntax can hide the actual reversal logic if over-used.
- **Quick check:** Keep guards simple; main loop is cleaner without extra let blocks.
- **Flash line:** "Guard early, then unwrap once."

### 7) Single-node early return is optional
- **Gotcha:** `if (head.next == null) return head` is not strictly necessary.
- **Quick check:** Main loop handles it fine; early return just saves one iteration.
- **Flash line:** "Optional optimization; correctness is the same."
