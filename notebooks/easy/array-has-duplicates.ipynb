## Problem Statement
Given a non-empty array of integers nums, every element appears twice except for one. Find that single one. You must implement a solution with a linear runtime complexity and use only constant extra space.

### Inputs
- nums: unsorted array of integers
### Output
- integer : unique non repeated integer

### Examples
- nums = [2,2,1] → 1
- nums = [4,1,2,1,2] → 4
- nums = [1] → 1

### Constraints
- 1 <= nums.length <= 3 * 10^4
- -3 * 10^4 <= nums [i] <= 3 * 10^4
- Each element in the array appears twice except for one element which appears only once.


## Approach #1 : XOR Operator

```kotlin
fun findUniqueNumber(nums : IntArray) : Int {
    println("Input: ${nums.contentToString()}")
    if (nums.size == 1) return nums[0]
    var ans = 0
    for(current in nums) {
        ans = ans xor current
    }
    return ans
}
println("${findUniqueNumber(intArrayOf(1))}")
println("${findUniqueNumber(intArrayOf(2,2,1))}")
println("${findUniqueNumber(intArrayOf(4,1,2,1,2))}")
```

### Developer Notes

#### Approach #1: Bit Manipulation (XOR)
- Initialize `ans = 0`.
- Iterate through the array once.
- For each value `num`, update: `ans = ans xor num`.
- Return `ans` at the end.

#### Why this works
- XOR rules:
  - `a xor a = 0` (same values cancel out)
  - `a xor 0 = a` (identity)
  - XOR is commutative and associative, so order does not matter.
- Since every number appears exactly twice except one, all pairs cancel and only the unique number remains.

#### Notes on implementation
- This is the optimal approach for the given constraints (linear time + constant extra space).
- No sorting, no hash map/set, no extra array needed.

## Time and Space Complexity

### Approach #1: XOR Accumulation
- **Time:** `O(n)`
  - One pass through the array.
- **Space:** `O(1)`
  - Uses only one accumulator variable (`ans`).

## Gotchas (Flash Cards)

### 1) Works only under exact frequency condition
- **Gotcha:** XOR approach assumes *every other element appears exactly twice*.
- **Quick check:** If elements can appear 3+ times, this logic is not directly valid.
- **Flash test:** `[2,2,2,3]` does **not** match problem assumptions.

### 2) Mislabeling the approach
- **Gotcha:** This is not a two-pointer technique.
- **Quick check:** Label it as **Bit Manipulation / XOR**.
- **Flash line:** "XOR accumulator, not pointers."

### 3) Don’t add unnecessary data structures
- **Gotcha:** Using `HashSet`/`Map` breaks constant-space requirement.
- **Quick check:** Keep only `ans`.
- **Flash line:** "Optimal = O(n) time, O(1) space."

### 4) Initialization matters
- **Gotcha:** `ans` must start at `0` (XOR identity).
- **Quick check:** `var ans = 0`.
- **Flash test:** `[1] -> 1`.

### 5) Negative numbers are safe
- **Gotcha:** Some people think XOR fails for negatives.
- **Quick check:** Bitwise XOR works fine with signed integers too.
- **Flash test:** `[-1, 2, 2] -> -1`.

### 6) Empty array handling
- **Gotcha:** Prompt says non-empty, but generic utility code may still guard.
- **Quick check:** either rely on constraints or add validation for production code.
- **Flash line:** "Interview constraints: non-empty."
