## Problem Statement
Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

### Inputs
- nums: unsorted array of integers
### Output
- true/false : true if any duplicates, false if every element is distinct

### Examples
- nums = [1,2,3,1] → true
- nums = [1,2,3,4] → false
- nums = [1,1,1,3,3,4,3,2,4,2] → true

### Constraints
- 1 <= nums.length <= 10^5
- 10^9 <= nums [i] <= 10^9


## Approach #1 : Two Pointers (Current & Next|Previous)

```kotlin
fun containsDuplicates(nums : IntArray) : Boolean {
    println("Input: ${nums.contentToString()}")
    if (nums.size < 2) return false
    nums.sort()
    for(current in 0 until nums.size-1) {
        if (nums[current] == nums[current+1]) return true
    }
    return false
}
println("${containsDuplicates(intArrayOf(1))}")
println("${containsDuplicates(intArrayOf(1,2,3,1))}")
println("${containsDuplicates(intArrayOf(1,2,3,4))}")
println("${containsDuplicates(intArrayOf(1,1,1,3,3,4,3,2,4,2))}")
```

## Approach #2 : HashSet

```kotlin
fun containsDuplicatesSet(nums : IntArray) : Boolean {
    println("Input: ${nums.contentToString()}")
    if (nums.size < 2) return false
    val set = mutableSetOf<Int>()
    for (num in nums) {
        val addded = set.add(num)
        if (!addded) return true
    }
    return false
}
println("${containsDuplicatesSet(intArrayOf(1))}")
println("${containsDuplicatesSet(intArrayOf(1,2,3,1))}")
println("${containsDuplicatesSet(intArrayOf(1,2,3,4))}")
println("${containsDuplicatesSet(intArrayOf(1,1,1,3,3,4,3,2,4,2))}")
```

### Developer Notes

#### Approach #1: Sort + Adjacent Scan
- Sort the array in-place (`nums.sort()`).
- After sorting, duplicates become adjacent.
- Scan once and compare neighbors:
  - if `nums[i] == nums[i + 1]`, return `true`
- If no adjacent match is found, return `false`.

#### Why this works
- Sorting groups equal values together.
- So checking only adjacent elements is sufficient to detect duplicates.

#### Important behavior
- This approach mutates the input array because `nums.sort()` is in-place.
- If original order must be preserved, sort a copy instead.

---

#### Approach #2: HashSet (Seen Values)
- Create an empty set.
- Iterate through the array once:
  - try adding each number into the set
  - if add fails, that value already exists → return `true`
- If loop completes, return `false`.

#### Why this works
- A set stores unique values only.
- Encountering an already-seen value immediately proves a duplicate.

#### Important behavior
- This approach does not need sorting and does not depend on order.
- It uses extra memory proportional to number of unique elements.

## Time and Space Complexity

### Approach #1: Sort + Adjacent Scan
- **Time:** `O(n log n)` (sorting dominates)
- **Space:** `O(1)` extra (algorithmic view; sort internals may use small stack/temporary memory)

### Approach #2: HashSet
- **Time:** `O(n)` average (each set add/lookup is average `O(1)`)
- **Space:** `O(n)` extra (set can store up to `n` elements)

### Tradeoff Summary
- Use **Sort + Scan** when minimizing extra memory is preferred.
- Use **HashSet** when aiming for faster average runtime.

## Gotchas (Flash Cards)

### 1) Single element / empty input
- **Gotcha:** Arrays with fewer than 2 elements cannot have duplicates.
- **Quick check:** return `false` when `nums.size < 2`.
- **Flash test:** `[1] -> false`

### 2) Current early return bug
- **Gotcha:** `if (nums.size < 2) return true` is incorrect for this problem.
- **Quick check:** should be `return false`.
- **Flash test:** `[42]` should not be duplicate.

### 3) Sorting changes input
- **Gotcha:** `nums.sort()` mutates the original array.
- **Quick check:** verify whether mutation is acceptable in the interview/problem.
- **Flash line:** "In-place sort = modified input."

### 4) Complexity is not O(n)
- **Gotcha:** Even though scan is linear, sorting is `O(n log n)`.
- **Quick check:** dominant step decides final complexity.
- **Flash test:** large `n` highlights sorting cost.

### 5) Adjacent compare boundary
- **Gotcha:** Loop should stop at `nums.size - 2` (or `until nums.size - 1`) to safely read `i + 1`.
- **Quick check:** avoid out-of-bounds at last index.
- **Flash line:** "Compare pairs `(i, i+1)` only."

### 6) Negative and large values are fine
- **Gotcha:** Value range does not affect correctness for sort+scan.
- **Quick check:** algorithm relies on equality, not value magnitude.
- **Flash test:** `[-1, 10^9, -1] -> true`

### 7) Alternative optimal tradeoff
- **Gotcha:** If asked for linear time, use `HashSet`.
- **Quick check:** Set approach gives `O(n)` time with `O(n)` space.
- **Flash line:** "Sort+scan saves extra space; Set saves time."
