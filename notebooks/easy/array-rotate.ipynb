## Problem Statement
Given an integer array nums, rotate the array to the right by k steps, where k is non-negative.

### Inputs
- nums: array of integers
- k : no of steps to rotate
### Output
- nums: shifted array of integers

### Examples
- nums = [1,2,3,4,5,6,7], k = 3 -> [5,6,7,1,2,3,4]
- nums = [-1,-100,3,99], k = 2 -> [3,99,-1,-100]

### Constraints
- 1 <= nums.length <= 10^5
- -2^31 <= nums[i] <= 2^31 - 1
- 0 <= k <= 105


## Approach #1 : Modulo Index

```kotlin
fun rotateArrayModulo(nums: IntArray, k: Int): IntArray {
    println("input : ${nums.contentToString()}")
    val size = nums.size
    if (size < 2) return nums
    val k = k % size
    val temp = IntArray(size)
    for (i in nums.indices) {
        temp[(i + k) % size] = nums[i]
    }
    return temp
}
println("Output : ${rotateArrayModulo(intArrayOf(1, 2, 3, 4, 5, 6, 7), 3).contentToString()}")
println("Output : ${rotateArrayModulo(intArrayOf(-1, -100, 3, 99), 2).contentToString()}")
```

## Approach#2 : Reverse sub arrays

```kotlin
fun rotateArrayReverse(nums: IntArray, k: Int): IntArray {
    println("input : ${nums.contentToString()}")
    val size = nums.size
    if (size < 2) return nums
    val k = k % size
    nums.reverse()
    nums.reverse(0, k)
    nums.reverse(k, size)
    return nums
}
println("Output : ${rotateArrayReverse(intArrayOf(1, 2, 3, 4, 5, 6, 7), 3).contentToString()}")
println("Output : ${rotateArrayReverse(intArrayOf(-1, -100, 3, 99), 2).contentToString()}")
```

### Developer Notes

#### Approach #1: Modulo Index (with temp array)
- Treat array indices as circular positions from `0` to `n-1`.
- For each element at index `i`, its rotated position after right shift by `k` is:
  - `newIndex = (i + k) % n`
- We place elements into a temporary array at `newIndex`.
- This approach is straightforward and easy to reason about, but uses extra memory.

#### Approach #2: Reverse Subarrays (in-place)
- First reduce rotations with `k = k % n`.
- Reverse the entire array.
- Reverse the first `k` elements.
- Reverse the remaining `n-k` elements.
- This works because full reverse brings the last `k` elements to the front (in reverse order), and the two partial reverses restore correct order.
- This is the preferred approach when extra array space is not allowed.

### Time and Space Complexity

#### Approach #1: Modulo Index
- **Time Complexity:** `O(n)`
  We visit each element once to place it in its new index.
- **Space Complexity:** `O(n)`
  A temporary array of size `n` is used.

#### Approach #2: Reverse Subarrays
- **Time Complexity:** `O(n)`
  Array is reversed in three linear passes.
- **Space Complexity:** `O(1)`
  Reversal is done in-place using constant extra variables.

### Real-world meaning of modulo
Think of `%` as wrap-around movement on a loop.
- Clock: after 12 comes 1 (wrap)
- Circular track: after last lane marker, you’re back at start
- Carousel seats: move forward k seats, you may cross the end and continue from seat 0

So `% n` answers:

> “If positions are only 0..n-1, where do I land after moving, even if I go past the end?”
---
### Rotate-array intuition (why modulo appears)
In rotate-right by k:
- array indices are a circle of size n
- each element moves k steps right
- if index crosses n-1, it wraps to start

So for an element currently at index i: [ newIndex = (i + k) % n ]

That formula is not random — it is exactly “move on a ring and wrap”.
Example: `n=7`, `i=5`, `k=3`
raw move: `5+3=8`
valid indices are `0..6`
wrapped index: `8 % 7` = `1`
So element at index 5 goes to index 1.

**Use these templates:**
- Forward move: (i + step) % n
- Backward move: (i - step + n) % n (add n to avoid negative index before %)

If array length is n, rotating by k is same as rotating by:
- rotate by n + 2 == rotate by 2
- rotate by 1000003 on n=7 is same as 1000003 % 7
- `k = k % n` (So if `k >= n`, reduce it first.)
